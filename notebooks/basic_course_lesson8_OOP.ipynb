Напишите программу, которая запрашивает у пользователя три целых числа и вычисляет среднее арифметическое этих чисел.

Подсказка:

Среднее арифметическое рассчитывается как сумма чисел, делённая на их количество.

***Как это задание выполнила бы я сама***



````python
def average_of_three(num1, num2, num3):
    return (num1 + num2 + num3) / 3

def num_input(num):
    return int(input(num))

num1 = num_input("Введите первое число: ")
num2 = num_input("Введите второе число: ")
num3 = num_input("Введите третье число: ")

print("Среднее арифметическое:", average_of_three(num1, num2, num3))
````

***Как я сделала это с помощью AI***

````python
def get_number(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Это не число, попробуйте ещё раз.")

def decimal_places(num):
    """Возвращает количество знаков после запятой у числа."""
    s = str(num)
    if '.' in s:
        return len(s.split('.')[1])
    return 0

def average_of_three(num1, num2, num3):
    result = (num1 + num2 + num3) / 3
    # Определяем максимальное количество знаков после запятой
    max_decimals = max(decimal_places(num1), decimal_places(num2), decimal_places(num3))
    return round(result, max_decimals)

# Ввод трёх чисел
num1 = get_number("Введите первое число: ")
num2 = get_number("Введите второе число: ")
num3 = get_number("Введите третье число: ")

# Вычисление и вывод среднего
print("Среднее арифметическое:", average_of_three(num1, num2, num3))
````

#### Задача 1: Переворот строки

Напишите программу, которая запрашивает у пользователя строку и выводит её в обратном порядке.

**Пример:**

```
Введите строку: Python
Результат: nohtyP
```

````python
def reverse_string(str):
  return str[::-1]

input_str = input("Введите строку:")
print ("Результат:",reverse_string(input_str))
````

#### Задача 2: Подсчёт слов в предложении

Запросите у пользователя предложение и подсчитайте количество слов в нём.

**Пример:**

```
Введите предложение: Это простое предложение
Количество слов: 3
```

````python
def count_words(sentence):
  words = sentence.split()
  return len(words)

input_sentence = str(input("Введите предложение:"))
print("Количество слов", count_words(input_sentence))
````

#### Задача 3: Замена символов

Запросите у пользователя строку и замените все пробелы на подчёркивания.

**Пример:**

```
Введите строку: Hello World Python
Результат: Hello_World_Python
```

````python
def change_space_on_underlining(str):
  return str.replace(" ","_")

input_str = str(input("Введите строку:"))
print("Результат:",change_space_on_underlining(input_str))
````
