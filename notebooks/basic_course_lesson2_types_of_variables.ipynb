**ЗАДАНИЕ**⃧

Напишите программу, которая запрашивает у пользователя три целых числа и вычисляет среднее арифметическое этих чисел.

Подсказка:

Среднее арифметическое рассчитывается как сумма чисел, делённая на их количество.

***Как это задание выполнила бы я сама***

```python
def average_of_three(num1, num2, num3):
    return (num1 + num2 + num3) / 3

def num_input(num):
    return int(input(num))

num1 = num_input("Введите первое число: ")
num2 = num_input("Введите второе число: ")
num3 = num_input("Введите третье число: ")

print("Среднее арифметическое:", average_of_three(num1, num2, num3))
```

***Как я сделала это с помощью AI***

```python
def get_number(prompt):
    while True:
        try:
            return float(input(prompt))
        except ValueError:
            print("Это не число, попробуйте ещё раз.")

def decimal_places(num):
    """Возвращает количество знаков после запятой у числа."""
    s = str(num)
    if '.' in s:
        return len(s.split('.')[1])
    return 0

def average_of_three(num1, num2, num3):
    result = (num1 + num2 + num3) / 3
    # Определяем максимальное количество знаков после запятой
    max_decimals = max(decimal_places(num1), decimal_places(num2), decimal_places(num3))
    return round(result, max_decimals)

# Ввод трёх чисел
num1 = get_number("Введите первое число: ")
num2 = get_number("Введите второе число: ")
num3 = get_number("Введите третье число: ")

# Вычисление и вывод среднего
print("Среднее арифметическое:", average_of_three(num1, num2, num3))
```
